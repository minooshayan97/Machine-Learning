```python
import numpy as np
import pandas as pd
```

# Preprocessing

```python
df=pd.read_csv('HW5.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

## dealing with missing values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df.iloc[:,:]=imputer.fit_transform(df)
```

```python
categorical_columns = ['WorkClass','MaritalStatus','Occupation','Relationship','Race','Sex','NativeCountry']
ordinal_columns = ['Education-num']
numerical_columns = ['Age','FinancialWeight','CapitalGain','CapitalLoss','HourPerWeek']
```

## encoding categorical features

```python
from sklearn.preprocessing import LabelEncoder
target_enc = LabelEncoder()
df['y'] = target_enc.fit_transform(df['Income'])
# <=50K is labeled 0
# >50K is labeled 1
```

```python
df.drop(columns=['Education'],inplace=True)
df.drop(columns=['Income'],inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first')
data = pd.DataFrame(encoder.fit_transform(df[categorical_columns]).toarray())
data.rename(columns={i:'x{}'.format(i) for i in range(0,len(data.columns))},inplace=True)
data = data.join(df[ordinal_columns])
data = data.join(df[numerical_columns])
data
```

# Baseline

```python
from sklearn.model_selection import cross_val_score
from sklearn.dummy import DummyClassifier
clfB = DummyClassifier(strategy = 'constant', constant = 0)
clfB_scores = cross_val_score(clfB, data, df['y'] ,cv=10)
np.mean(clfB_scores)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

clf1 = LogisticRegression()
clf1_scores = cross_val_score(clf1, data, df['y'] ,cv=10)
np.mean(clf1_scores)
```

```python
clf1.get_params()
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

clf2 = GaussianNB()
clf2_scores = cross_val_score(clf2, data, df['y'] ,cv=10)
np.mean(clf2_scores)
```
